```python
import pandas as pd
# <--- Đọc dữ liệu sheet Task --->
def read_Task(excel_file_path):
    # Đọc dữ liệu Task
    task_df = pd.read_excel(excel_file_path, sheet_name='Task', nrows=41)

    # Đọc dữ liệu từng cột sheet Task
    task_id_list = task_df['ID'].tolist()
    task_name_list = task_df['Name'].tolist()
    task_group_list = task_df['Group'].tolist()
    task_predecessors_list = task_df['Predecessors'].tolist()
    task_duration_list = task_df['Duration'].tolist()
    task_resource_requirement_list = task_df['Resource Requirement'].tolist()
    

    #<--- Hàm chuẩn hóa dữ liệu resource_requirement để tìm ra resource phù hợp với task --->
    resource_require_list = dict()
    for index, row in task_df.iterrows():
        id = int (row['ID'])
        resource_require_list[id] = dict()

        task = row['Resource Requirement']
        idx = task.find(':')
        if idx != -1:
            role = task[:idx].strip()
            resource_require_list[id]['role'] = role
            resource_require_list[id]['skill'] = dict()
            for text in task [idx + 1:].split(','):
                skill, num = text.strip().split(':')
                skill = skill.strip()
                num = int(num.strip())
                resource_require_list[id]['skill'][skill] = num
    # print(resource_require_list)
    return task_id_list, task_name_list, task_duration_list, resource_require_list

# <--- Đọc dữ liệu sheet Resource --->
def read_Resources(excel_file_path):
    #Đọc dữ liệu Resource
    resource_df = pd.read_excel(excel_file_path, sheet_name = 'Resources', nrows = 20)

    #Đọc dữ liệu từng cột sheet Resources
    resource_id_list = resource_df['ID'].tolist()
    resource_type_list = resource_df['Type'].tolist()
    resource_description_list = resource_df['Description'].tolist()
    resource_qualities_list = resource_df['Qualities'].tolist()
    resource_skill_groups_list = resource_df['Skill Groups'].tolist()
    resource_cost_list = resource_df['Cost'].tolist()

    #<--- Hàm chuẩn hóa dữ liệu resource_qualities để tìm ra resource phù hợp với từng task --->
    qualities_list = dict()
    for index, row in resource_df.iterrows():
        id = int (row['ID'])
        qualities_list[id] = dict()

        resource = row['Qualities']
        idx = row['Type']
        qualities_list[id]['Type'] = idx
        qualities_list[id]['skill'] = dict()
        for text in resource.split(','):
            skill, num = text.strip().split(':')
            skill = skill.strip()
            num = int(num.strip())
            qualities_list[id]['skill'][skill] = num
    # print(qualities_list)

    return resource_id_list, resource_type_list, resource_description_list, qualities_list, resource_skill_groups_list, resource_cost_list

excel_file_path = 'Data.xlsx'

# Đọc dữ liệu từ sheet Task
id_col, name_col, duration_list, resource_require_list= read_Task(excel_file_path)

# Đọc dữ liệu từ sheet Resources
rs_id, re_type, rs_des, rs_qualities, rs_skgroups, rs_cost = read_Resources(excel_file_path)
```

```python
import numpy as np
staff = {}
time_staff = {}
for i in range(1, 21):
  staff[i] = list() # list thời gian staff[i][x] là số giờ được đánh 1 ở trong sheet Timeframe, kết quả trả ra là giờ kết thúc của timeslot x được đánh '1'
  staff[i].append(pd.Timestamp('2023-01-02 08:00:00'))
  time_staff[i] = dict() # từ thời gian -> duration (  Ngược lại của staff[]  )

#<----- Phần đọc dữ liệu Timeframe ----->
def read_Timeframe(excel_file_path):
  time = np.zeros(21)
  df = pd.read_excel(excel_file_path, sheet_name = 'Timeframe')
  for index, row in df.iterrows(): #đọc theo hàng
    for i in range(1, 21):
      if row[f'ID_{i}'] == 1: # cột có giá trị 1
        for j in range(0, 4):
          staff[i].append(row['end'] - pd.Timedelta(hours=(3-j))) # Thời gian ghi vào list staff = (Tăng dần 1 tiếng từ thời gian bắt đầu làm đến khi kết thúc phiên làm)
      delta_t = pd.Timedelta(hours=1) # Delta_t = 1 tiếng
      t_curr = row['start'] + delta_t #Thời gian hiện tại bằng thời gian bắt đầu + 1 tiếng

      while t_curr <= row['end']: #Nếu thời gian hiện tại chưa bé hơn thời gian kết thúc
        if row[f'ID_{i}'] == 1:  time[i] += 1 
        time_staff[i][t_curr] = time[i]
        t_curr = t_curr + delta_t

read_Timeframe('Data.xlsx')
```

```python
import numpy as np
# <--- Hàm tìm người phù hợp cho từng công việc --->
def Find_Source_Fit(task_id):
  task_requirements = resource_require_list[task_id]
  suitable_resources = [] # list giao việc cho nhân viên ([22, 5]: Task có id 22 được giao cho người có id 5)
  for i in rs_qualities:
      if rs_qualities[i]['Type'] == task_requirements['role']: # so sánh role người được yêu cầu làm task với type của nhân viên
          res_qualities = rs_qualities[i]['skill'] # list skill nhân viên có type == role
          task_req = task_requirements['skill'] # list skill của role được yêu cầu trong task
          cnt = 0 # biến đếm để kiểm tra xem đã check hết số lượng skill của role được yêu cầu trong task chưa
          for key in task_req:
              if res_qualities[key] >= task_req[key]: #check cấp độ kĩ năng của nhân viên phải >= cấp độ kỹ năng của role đươc yêu cầu
                  cnt += 1
          if cnt == len(task_req):
              suitable_resources.append(i) # cập nhân nhân viên đó trong list giao việc

  return suitable_resources
```

```python
#<--- Hàm xếp lịch vào timeframe và ghi kết quả --->
work = np.zeros(21) # việc mà resource ID đang làm
complete = [0] # việc đã hoàn thành
viecCoTheLam = [] # danh sách chứa các công việc có thể làm tại thời diểm t
viecDangLam = [] # danh sách chứa các công việc đang làm tại thời điểm t

clock = dict() # map lưu trữ: thời gian mà nhân viên có id vừa hoàn thành task được giao
t = pd.Timestamp('2023-01-02 08:00:00') # biến đếm thời gian
delta_t = pd.Timedelta(hours=1) # delta_t = 1 tiếng

#Tính thời gian nhân viên làm
work_time = np.zeros(21) # Mảng lưu trữ thời gian nhân viên làm việc trong morning shift và afternoon shirt
work_overtime = np.zeros(21) # Mảng lưu trữ thời gian nhân viên làm việc trong over time

#Những việc nhân viên làm
list_work = dict() # map lưu trữ những việc mà nhân viên đã làm
for id in range(1, 21):
  list_work[id] = list()

#<----- Hàm tìm thời gian bắt đầu của task ----->
def compute(t, d, id): 
  _output_start.append(t) # Lưu kết quả thời gian bắt đầu của task vào file kết quả
  while d > 0:
    try:
      if time_staff[id][t + delta_t] - time_staff[id][t] == 1: #từ t đến t+1 là nhân viên id có ca làm việc
        d -= 1 # cho nhân viên id làm việc
        if t.hour >= 17 and t.hour < 21:
          work_overtime[id] += 1 #Tính thời gian over_time của nhân viên có id
        else:
          work_time[id] += 1 #Tính thời gian làm việc morning shift và morning shift của nhân viên có id
    except:
      pass
    t += delta_t # tăng t lên 1 tiếng
  return t # trả ra thời gian hoàn thành

 #<--- Hàm phân công task cho resource --->
def Phan_Cong(t, task):
  if len(Find_Source_Fit(task) ) == 0: #Task không có resource phù hợp để làm -> không xếp 
    #print(f'TASK {task} khong tim duoc nguoi thuc hiện')
    if task not in complete: #Nếu task không có trong danh sách task đã hoàn thành
      complete.append(task) #add việc vào danh sách đã hoàn thành
      _output_id_task.append(task) #add id task vào kết quả
      _output_id_resource.append('None') #add id resource = None vào kết quả
      _output_start.insert(len(_output_id_task) - 1, 'None') #add thời gian bắt đầu bằng None vào kết quả
      _output_end.append('None') #add thời gian kết thúc = None vào kết quả
    viecKhongTheLam.append(task) #Thêm task vào danh sách việc không thể làm được
    

  for id in Find_Source_Fit(task): #Nếu tìm được resource phù hợp
    if work[id] == 0: # kiểm tra resource có id có đang làm việc gì không
      work[id] = task # giao task cho người có id đó
      d = int(duration_list[task-1]) # d = duration của task đó
      viecDangLam.append(task) # add task đó vào danh sách việc đang làm
      new_t = compute(t, d, id)  # tính toán deadline mà resource có id nhận để làm xong task
      # print(f'--> Đặt clock cho nhân viên {id}: {task} - {new_t}') 
      try:
        clock[new_t].append(id) # Nếu đã có list clock(deadline) cho id thì thêm vào
      except:
        clock[new_t] = [id,] #Chưa có thì tạo rồi thêm id vào
      break; #thoát vàng for

# <--- Tạo danh sách để lưu kết quả
_output_id_task = [] #id task
_output_id_resource = [] #id resource được giao làm task
_output_start = [] #thời gian bắt đầu của task
_output_end = [] #thời gian kết thúc của task

viecCoTheLam = [] #Danh sách các task có thể làm tại thời điểm t
viecKhongTheLam = [] #Danh sách các task không thể làm

begin = True
cnt = 0

while True:
  # Kiểm tra clock xác định việc đã hoàn thành
  check = False # biến kiểm tra xem có việc nào hoàn thành không
  try:
    for id in clock[t]: #danh sách clock theo t
        _output_id_resource.append(int(id)) #Thêm id_resource của việc đã hoàn thành vào kết quả
        _output_id_task.append(work[int(id)]) #Thêm id task (work[int(id)]: Việc mà người có id đang làm) đã hoàn thành vào kết quả

        # _output_start.append() là thời gian đặt clock trong hàm compute
        _output_end.append(t) #Thêm thời gian t là thời gian kết thúc của task

        complete.append(work[int(id)]) #thêm task đó vào danh sách việc đã hoàn thành
        list_work[int(id)].append(work[int(id)]) #thêm task vào danh sách việc đã làm của người có resource id
        work[int(id)] = 0 #việc đang làm của người đó reset về 0
    # print(f'clock {t}: {complete}') #Hiển thị thời gian hoàn thành task
    check = True #Việc đã hoàn thành
  except:
    pass

  if check or begin: #Nếu việc chưa hoàn thành hoặc chưa bắt đầu
    begin = False #bắt đầu = false
    viecCoTheLam = [] #tạo danh sách việc có thể làm mới
    for i in range(1, 42):
      if i not in complete and i not in viecDangLam and all(element in complete for element in pre_list[i]): #Nếu việc không nằm trong danh sách đã làm, đang làm hoặc nằm trong danh sách pre_list của list khác
        viecCoTheLam.append(i) #thêm việc có id là i vào danh sách việc có thể làm
    # print(viecCoTheLam)

  # xếp lịch cho những việc có thể làm tức là đặt clock cho việc đó
  for task in viecCoTheLam:
    Phan_Cong(t, task)

  t += delta_t #tăng t lên 1 tiếng
  # print(t)

  if len(set(complete)) == 42: #Đã hoàn thành xếp thời gian cho tất cả việc
    break

```
